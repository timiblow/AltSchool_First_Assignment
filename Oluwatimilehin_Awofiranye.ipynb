```python
# Importing necessary libraries

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

## Task One

```python
# Load 'Energy' data explicitly specify the engine for .xls files
# Removing first two columns
# Changing column labels to preferred labels
# Replacing country with '...' with NaN


Energy = pd.read_excel('Energy Indicators.xls', 
                       engine='xlrd', 
                       names = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'], 
                       header=16, 
                       skiprows=[0], 
                       usecols=[2,3,4,5], 
                       skipfooter=38, 
                       na_values="...")

Energy.head()
```

```python
#  Converting 'Energy Supply' from gigajoules to petajoules
Energy['Energy Supply'] = Energy['Energy Supply'] * 1_000_000
```

```python
# Load the data after conversion
Energy
```

### Removing unnecessary additional objects from country names in the dataset

```python
# replacing numbers at the end of the string with an empty string
Energy['Country'] = Energy['Country'].str.replace(r"\d+", "", regex=True).unique()

# replace parenthesis and its contents with an empty string
Energy['Country'] = Energy['Country'].str.replace(r"\W+\(.*\)", "", regex=True)
```

```python
# Load 'Country' column to see if changes has been made to country names
Energy['Country'].unique()
```

```python
# Replacing specific country names with preferred country names

Energy['Country'] = Energy['Country'].replace({
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
})
```

## Task Two

```python
# Load dataset

GDP = pd.read_csv('world_bank.csv', skiprows=4)
GDP.head()
```

```python
# Replacing specific country names with preferred country names

GDP['Country Name'] = GDP['Country Name'].replace({
    "Korea, Rep.": "South Korea", 
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"
})
```

```python
# Load 'Country Name' column to see changes

GDP['Country Name'].unique()
```

## Task Three

```python
# Load 'SCIMAGOJR' data and explicitly specify the engine for .xls files

ScimEn = pd.read_excel('scimagojr-3.xlsx', engine='openpyxl')
ScimEn
```

```python
# Set 'Country' as index for the dataset

ScimEn.set_index('Country', inplace=True)
```

```python
# Load data

ScimEn
```

```python
# Mapping out GDP from the last 10 years(2006-2015)

yrs = []
for i in range(2006,2016):
    yrs.append(str(i))
```

```python
# creating a DataFrame for GDP from the last 10 years with their Country Names
GDP = GDP[['Country Name', *yrs]]
```

### Combining the three datasets (Energy, GDP, ScimEn) altogether into one DataFrame

```python
# Set Country as index for data in 'Energy' dataset
Energy = Energy.set_index('Country')
```

```python
# Set 'Country Name' as index for data in 'Energy' dataset
bind = GDP.set_index('Country Name')
```

```python
# Join 'energy' and 'ScimEn' datasets by the intersection of 'Country Names'
bind_df = ScimEn.join(Energy, 'Country')
```

''' 
Since the task is to combine the three datasets by the intersection of their Country Names and after joining both 'energy' and 'ScimEn' datasets into one DataFrame assigned to 'bind_df, we would now join them with 'GDP' dataset and combine all three datasets into one DataFrame 
'''

```python
# Combining all three datasets into one DataFrame using the intersection of country names

df = bind_df.join(bind, 'Country')
```

```python
# Reducing data entries to top 15 countries by method of masking

mask = df['Rank'] <= 15
```
